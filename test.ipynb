# Open webcam

```python
import cv2
from deepface import DeepFace

cam = cv2.VideoCapture(0)

frame_width = int(cam.get(cv2.CAP_PROP_FRAME_WIDTH)) #640
frame_height = int(cam.get(cv2.CAP_PROP_FRAME_HEIGHT)) #480

while True:
    ret, frame = cam.read()

    # Ret is a boolean that indicates wether the frame worked or not
    if not ret:
        break
    #print(frame)

    cv2.imshow('image', frame)

    c = cv2.waitKey(10) 
    # Press a to close
    if c == 97:
        break
    
    result = DeepFace.analyze(frame, actions="emotion", enforce_detection=False)
    #print(result)
    
    result = result[0]
    emotion = result.get('dominant_emotion')
    confidence = result.get('face_confidence')
    print(emotion, confidence)


cam.release()
cv2.destroyAllWindows()
```
