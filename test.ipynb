```python
import cv2

cam = cv2.VideoCapture(0)

while True:
    ret, frame = cam.read()

    # ret is a boolean that indicates wether the frame worked or not
    if not ret:
        break
    print(frame)

    cv2.imshow('image', frame)

    c = cv2.waitKey(10) 
    # press a to close
    if c == 97:
        break
    



cam.release()
cv2.destroyAllWindows()
```
